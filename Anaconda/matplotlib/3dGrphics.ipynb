```python
import matplotlib        as mlt
import matplotlib.pyplot as plt
import numpy             as np
```

# **Graphiques en 3D**

### **Les nuages de points**

```python
x = np.random.normal(0,1,100) # moyenne, Ã©cart type, nombre de valeurs
y = np.random.normal(0,1,100)
z = np.random.normal(0,1,100)
```

```python
# la rendre deplaÃ§able `%matplotlib`
%matplotlib
plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
ax.scatter3D(x,y,z, s=40) # s=taille dees points

# placer les axes
ax.set_xlabel('axe x')
ax.set_ylabel('axe y')
ax.set_zlabel('axe z')
```

### **Courbe en 3D**

```python
val_x = np.linspace(0,20,200) # 200 Ã©lÃ©ments entre 0 et 20
val_y = np.cos(2*val_x)
val_z = np.sin(2*val_x)

plt.figure(figsize=(10,8))
ax1 = plt.axes(projection = '3d')
ax1.plot3D(val_x,val_y,val_z, lw=5, color='#FF6D2D') # Ã©paisseur de la linge orange 5
```
