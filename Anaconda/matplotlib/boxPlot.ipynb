# **Les diagrammes à boîtes (boîtes à moustaches)**

C'est utile pour vérifier la symétrie, la dispersion ou la centralité de la distribution des valeurs associées à une variable,  
C'est également utile à la comparaison de variable à des échelles analogues, ainsi qu'à la comparaison de valeurs, observation de groupes d'individus sur la même variable.

```python
data = np.random.normal(0,5,1000)
print(data)
```

```python
data        = np.random.normal(0,5,1000)

fig,ax      = plt.subplots()
ax.set_title('Box plot basic')
flierdesign = dict(marker='o',markerfacecolor='pink',markersize=8,markeredgecolor='gray')
ax.boxplot(data,flierprops=flierdesign) # Couleur des points
```

```python
data        = np.random.normal(0,5,1000)

fig,ax      = plt.subplots()
ax.set_title('Box plot basic')
flierdesign = dict(marker='o',markerfacecolor='pink',markersize=8,markeredgecolor='gray')
ax.boxplot(data,showfliers=False) # Sans les points
```

```python
data        = np.random.normal(0,5,1000)

fig,ax      = plt.subplots()
ax.set_title('Box plot basic')
flierdesign = dict(marker='o',markerfacecolor='orange',markersize=8,markeredgecolor='gray')
ax.boxplot(data,vert=False,showfliers=flierdesign) # Horizontale
```

```python
data1 = [np.random.normal(0,std,1000) for std in range(0,4)]

fig2,ax2 = plt.subplots(figsize=(12,12))
ax2.set_title('Box plot basic')
flierdesign = dict(marker='o',markerfacecolor='orange',markersize=8,markeredgecolor='gray')
ax2.boxplot(data1,showfliers=flierdesign)
```
