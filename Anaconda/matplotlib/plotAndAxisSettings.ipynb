```python
import matplotlib        as mlt
import matplotlib.pyplot as plt
import numpy             as np
```

# **Réglage du tracé et des axes**

```python
x=np.linspace(-20,20,100)
x
```

```python
plt.plot(x,np.cos(x)) # Les axes sont choisis par défaut
```

```python
plt.plot(x,np.cos(x))
# afficher l'axe de -1 à 15
plt.xlim(-1,15)
```

```python
plt.plot(x,np.cos(x))
plt.xlim(-5,15)
# afficher l'axe de -1.5 à 1.5
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.sin(x))
# afficher l'axe de 15 à -5, les ordonnées de -2 à 2
plt.axis([-2,15 , -2,2])
```

```python
plt.plot(x,np.sin(x))
# obtenir un repère orthonormé
plt.axis('equal')
plt.grid(True)
```

```python
plt.plot(x,np.sin(x))
# n'obtenir que la portion de la courbe
plt.axis('scaled')
plt.grid(True)
```

```python
plt.plot(x,np.sin(x))
# sans les axes
plt.axis('off')
```

```python
plt.plot(x,np.sin(x))
# sans les axes
plt.axis('on')
```
