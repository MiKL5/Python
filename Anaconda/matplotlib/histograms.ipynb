```python
age = [1,1,2,3,3,5,7,8,9,10,10,11,11,13,13,15,16,17,18,18,18,19,20,21,21,23,24,24,25,25,25,25,26,26,27,27,27,27,27,28,28,28,29,30,30,31,32,33,34,34,34,35,36,37,37,38,38,39,40,41,41,12,43,44,45,45,46,47,48,48,49,50,51,52,53,54,55,55,56,57,58,60,61,63,64,65,66,68,70,71,72,74,75,77,81,83,84,91]

plt.figure(figsize=(12,8))

plt.hist(age)
```

```python
plt.figure(figsize=(12,8))

plt.hist(age,bins=20) # bins : tranches
```

```python
plt.figure(figsize=(12,8))
plt.hist(age,bins=100)
```

```python
plt.figure(figsize=(12,8))

plt.hist(age,bins=[0,15,30,40,60,80,90,100]) # choix des tranches
```

```python
plt.figure(figsize=(12,8))
plt.hist(age,bins=10,color='b',histtype='barstacked')
```

```python
plt.figure(figsize=(12,8))
plt.hist(age,bins=10,color='b',histtype='barstacked',edgecolor='black') # edgecolor : conteur
```

```python
plt.figure(figsize=(12,8))
plt.hist(age,bins=10,color='b',histtype='barstacked',alpha=.5) # alpha = transparence
```

```python
plt.figure(figsize=(12,8))
plt.hist(age,bins=10,color='b',histtype='barstacked',alpha=.5,rwidth=.5) # rwidth = largeur des barres
```
