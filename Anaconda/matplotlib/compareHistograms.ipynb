```python
import matplotlib        as mlt
import matplotlib.pyplot as plt
import numpy             as np
```

# **Les histogrammes**

```python
# 3 distribution (normal et différents écarts types
x  =np.random.normal( 0,.8,1000)
x1 =np.random.normal(-3, 1,1000)
x2 =np.random.normal( 4, 2,1000)
```

```python
plt.figure(figsize=(10,8))
plt.hist(x ,label='distrib1',histtype='stepfilled',alpha=.5,bins=50)
plt.hist(x1,label='distrib2',histtype='stepfilled',alpha=.5,bins=50)
plt.hist(x2,label='distrib3',histtype='stepfilled',alpha=.5,bins=50)
plt.legend()
```

Il est possible de stocker les données dans un dictionnaire appeler au traçage de l'histogramme.

```python
x  =np.random.normal( 0,.8,1000)
x1 =np.random.normal(-3, 1,1000)
x2 =np.random.normal( 4, 2,1000)
style = dict(histtype='stepfilled',alpha=.5,bins=50)

plt.figure(figsize=(10,8))
plt.hist(x ,label='distrib1',**style)
plt.hist(x1,label='distrib2',**style)
plt.hist(x2,label='distrib3',**style)
plt.legend()
```

Avec Python, quand le dictionnaire est l'argument d'une fonction, il est précédé de `**`.
