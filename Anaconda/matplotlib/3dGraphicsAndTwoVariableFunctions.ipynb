# **Graphique en 3D d'une fonction à deux variables**

```python
def f(x,y):
    return np.cos(x)+np.sin(y)

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x,y = np.meshgrid(x,y)
z   = f(x,y)

plt.figure(figsize = (12,8))
ax = plt.axes(projection = '3d')
ax.plot_surface(x,y,z, cmap = 'plasma')

ax.set_xlabel('axe x')
ax.set_ylabel('axe y')
ax.set_zlabel('axe z')
```

```python
%matplotlib
def h(x,y):
    return 10- (x**2 + y**2)

plt.figure(figsize = (12,8))
X   = np.arange(-5,5, .2)
Y   = np.arange(-5,5, .2)
X,Y = np.meshgrid(X,Y)
Z   = h(X,Y)

# plt.scatter(X,Y)
ax1 = plt.axes(projection = '3d')
ax1.plot_surface(X,Y,Z, cmap = 'viridis')

ax1.set_xlabel('axe X')
ax1.set_ylabel('axe y')
ax1.set_zlabel('axe z')
```

```python
%matplotlib
def h(x,y):
    return 10- (x**2 + y**2)

plt.figure(figsize = (12,8))
X   = np.arange(-5,5, .2)
Y   = np.arange(-5,5, .2)
X,Y = np.meshgrid(X,Y)
Z   = h(X,Y)

# plt.scatter(X,Y)
ax1 = plt.axes(projection = '3d')
ax1.plot_wireframe(X,Y,Z, cmap = 'viridis') # Tracé filaire

ax1.set_xlabel('axe X')
ax1.set_ylabel('axe y')
ax1.set_zlabel('axe z')
```
