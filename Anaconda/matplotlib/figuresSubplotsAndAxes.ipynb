```python
import matplotlib        as mlt
import matplotlib.pyplot as plt
import numpy             as np
```

# **Figures, subplot et axes**

Les subplots sont des sous-parcelles.

`axis` et `axes` sont différents.

```python
fig = plt.figure()
fig.set_facecolor('r')
ax  = plt.axes()
ax1 = plt.axes([.5,.6,.25,.25]) # à 50% de la largeur, 60% de la hauteur, mesure 25% de la largeure et la hauteur de la figure
```

## **Créer plusieurs graphiques dans la même figure**
Ces graphhiques sont des sous-parcelles.

```python
fig = plt.figure()
ax  = fig.add_axes([.1,0.5,.8,.4]) # placée au début et la motiée de la heuteur de la figure, puis .8 pour 80% de la figure et .4 la hauteur
ax1 = fig.add_axes([.1,0,.8,.4])   # placé au début et bas de la figure, mêmes largeurs et hauteurs
```

```python
fig = plt.figure()
ax  = fig.add_axes([.1,0.5,.8,.4])
ax1 = fig.add_axes([.1,0,.8,.4])
# mettre le sinus dans l'axe 1 et le cosinus au second
x   = np.linspace(0,10)
ax .plot(np.sin(x))
ax1.plot(np.cos(x))
```

```python
# idem avec une boucle
for i in range(1,10):
    plt.subplot(3,3,i) # 3 lignes et colonnes et parcelles
```

```python
# Plus de place au percelles
fig = plt.figure(figsize = (10,8)) # en pouces

# Espacer les parcelle
fig.subplots_adjust(hspace=0.5,wspace=0.4)

for i in range (1,7):
    plt.subplot(2,3,i)
    plt.text   (.5,.5,i,fontsize=16) # indice au milieu des parcelles et taille à 16px
```

```python
# Plusieurs parcelles sans boucle `subplots`
fig,ax = plt.subplots(3,2)
```

```python
# Plusieurs parcelles sans boucle `subplots` et partager les axes `sarex` et `sharey`
fig,ax = plt.subplots(3,2,sharex = True, sharey = True)
```

```python
# Plusieurs parcelles sans boucle `subplots` et partager les axes avec `col` et `row`
fig,ax = plt.subplots(3,2,sharex = 'col', sharey = 'row')
```

```python
# Remplir les parcelles
fig,ax = plt.subplots(3,2,sharex = True, sharey = True)
ax[0,0].plot(x)
ax[0,1].plot(x**2)
ax[1,0].plot(np.sin(x))
ax[1,1].plot(np.cos(x))
ax[2,0].plot(np.tan(x))
ax[2,1].plot(np.e**x)
```
