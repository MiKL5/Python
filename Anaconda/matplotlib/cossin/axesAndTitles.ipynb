```python
import matplotlib        as mlt
import matplotlib.pyplot as plt
import numpy             as np
```

# **Axes et titres**

```python
x=np.linspace(0,20,100)
x
```

```python
# Cr√©er un graphique en ligne
plt.plot(x,x,linestyle='-')
```

```python
fig, ax = plt.subplots()

# Tracer la courbe en orange fluo
plt.plot(x, x, linestyle='-', color='#FF5500', linewidth=2)  # Orange
plt.plot(x, x+2, linestyle='--', color='#FFFF00', linewidth=2)  # Jaune
plt.plot(x, x+4, linestyle='-.', color='#00FF00', linewidth=2)  # Vert
plt.plot(x, x+6, linestyle=':', color='#9D00FF', linewidth=2)  # Violet

# Ajout d'une grille
plt.grid(True)

#¬†Nommer le graphique et les axes
plt.title('Repr√©sentation des fonctions affines',color='#FFF')
plt.xlabel('Abcsicces')
plt.ylabel('Ordonn√©es')

# Changer la couleur de fond de la figure
fig.patch.set_facecolor('black')

# Changer la couleur de fond des axes
ax.set_facecolor('black')

# Changer la couleur des ticks et des labels pour qu'ils soient visibles sur fond noir
ax.tick_params(colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

# Changer la couleur de la bordure des axes
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')
```

#### **Repr√©sentation des sinus et cosinus**

```python
x=np.linspace(-20,20,100)
x
```

```python
plt.plot(x,np.cos(x),'-',label='cosinus')
plt.plot(x,np.sin(x),':',label='sinus')

plt.title("Cosinus et sinus")
plt.xlabel('Abcsicces')
plt.ylabel('Ordonn√©es')
plt.legend()
plt.savefig('cossin.jpg',dpi=200)
plt.savefig('cossin.pdf',dpi=200)
plt.savefig('cossin',dpi=200)
plt.savefig('cossin.svg')
```
