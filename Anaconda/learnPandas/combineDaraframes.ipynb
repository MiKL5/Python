```python
import pandas as pd
```

# **Combine dataframes**

```python
data_villes = {
    'Paris'    :[75000,2175601,321,392,177,181],
    'Marseille':[13000, 868277,219,299, 99, 54],
    'Lyon'     :[69000, 518635, 78, 90, 26, 17]
}

# Les colonnes sont les villes, et les lignes sont les cat√©gories.
villes1   = pd.DataFrame(data_villes,
                        index=['code postal','population','ecole maternelle','ecole elementaire','college','lycee']
)
villes1
```

```python
data_villes2 = {
                'Avignon' :[84000, 92130,26,23037],
                'Bordeaux':[33000,249212,59,75974]
                }

villes2 = pd.DataFrame(data_villes2,index=['code postal', 'population', 'ecole maternelle', 'nombre eleves'])
villes2
```

```python
# Combiner les dataframes
villes = pd.concat([villes1,villes2])
villes    # Des ligne sont en douvle
```

```python
villes = pd.concat([villes1,villes2],axis=1)
villes
```

```python
# Que les lignes communes
villes_donnees_communes = pd.concat([villes1,villes2],axis=1,join='inner')
villes_donnees_communes
```
