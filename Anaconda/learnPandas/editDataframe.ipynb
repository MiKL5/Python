# **Modifier un Dataframe**

```python
notes_français = pd.Series({"Christophe": 0,  "Carla": 5, "Caroine": 20, "Jean-Baptiste": 12, "Florence": 19,   "Emma": 19})
notes_histoire = pd.Series({"Christophe": 0,  "Carla": 9, "Caroine": 20, "Jean-Baptiste": 18, "Florence": 20,   "Emma": 19})
notes_maths    = pd.Series({"Christophe": 17, "Carla": 6, "Caroine": 20, "Jean-Baptiste": 2,  "Florence": 15.5, "Emma": 19})
notes          = pd.DataFrame({'Français':notes_français,'Histoire':notes_histoire,'Maths':notes_maths})
notes
```

```python
notes['Moyenne'] = (notes['Français'] + notes['Histoire'] + notes['Maths'])/3
notes
```

```python
notes.round().astype(int)
```

```python
# Séléctionner la moyenne
notes[['Moyenne']].round().astype(int)
```

Quand utiliser inplace ?

inplace=True :  
Utile pour modifier le DataFrame sans créer de nouvelle variable.  
E.g. : df.drop(), df.fillna(), df.sort_values().

inplace=False (par défaut) :  
Utile pour conserver l’original et travailler sur une copie modifiée.  
E.g. : df_modifie = df.drop().

Méthodes courantes avec inplace

df.drop()              `df.drop('B', axis=1, inplace=True)`  
df.fillna()            `df.fillna(0, inplace=True)`  
df.sort_values()       `df.sort_values('A', inplace=True)`  
df.rename()            `df.rename(columns={'A': 'NouvelleColonne'}, inplace=True)`  
df.replace()           `df.replace(1, 100, inplace=True)`  
df.reset()             `df.reset_index(drop=True, inplace=True)`  
df.drop_duplicates()   `df.drop_duplicates(inplace=True)`  
df.set_index()         `df.set_index('A', inplace=True)`

Quand utiliser inplace ?  
Avantages :  
Évite de réaffecter le DataFrame (df = df.method()).  
Utile pour les chaînes de traitement où tu ne veux pas créer de copies intermédiaires.  
Inconvénients :  
* Peut rendre le code moins lisible (modification "invisible" du DataFrame).  
* Certaines méthodes ne supportent pas inplace (comme apply, astype, T).  

Bonnes pratiques  
* Évite inplace pour un code plus clair et plus facile à déboguer.  
* Préfère la réaffectation pour les opérations complexes ou en chaîne.
