```python
import numpy   as np
import pandas  as pd
import seaborn as sns
```

# **Multivariables : croisement de trois variables**

```python
student = pd.read_csv('student.csv')
student.head()
```

```python
sns.catplot(x = 'Fjob', y = 'Note3', hue = 'sex', data = student, kind = 'bar')
```

```python
sns.catplot(x = 'Fjob', y = 'Note3', hue = 'sex', data = student, kind = 'box')
```

```python
sns.catplot(x = 'Fjob', y = 'Note3', hue = 'sex', data = student, kind = 'violin')
```

```python
sns.catplot(x = 'Fjob', y = 'Note3', hue = 'sex', data = student, kind = 'violin', split = True)
```
