```python
import numpy      as np
import pandas     as pd
import seaborn    as sns
import matplotlib as mlt
```

# **Régressions linéaires**

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris['variety'].unique()
```

```python
# Proche de 1, la corrélation est forte.
correlation = iris.select_dtypes(include='number').corr()
correlation
```

```python
# Ajuster (tracer) les pétales et sépales dans uen diagrammes de régression linéaire
sns.lmplot(x = 'sepal.length', y = 'petal.length', data = iris)
```

```python
# Croiser avec les variétés, et les nuages plus grands `scatter_kws`
sns.lmplot(x = 'sepal.length', y = 'petal.length', data = iris, hue = 'variety', markers = ['o', 'p', 'v'], scatter_kws = {'s': 100})
```

Pour le "Setosa", la longueur des sépales augmente, mais pas pour les pétales.  
Pour les autres variétés, c'est positif.

```python
# Obtenir 3 graphiques `col`
sns.lmplot(x = 'sepal.length', y = 'petal.length', data = iris, col = 'variety')
```
