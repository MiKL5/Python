```python
import numpy      as np
import pandas     as pd
import seaborn    as sns
import matplotlib as mlt
```

# **Corrélations de variables**

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris['variety'].unique()
```

```python
# Proche de 1, la corrélation est forte.
correlation = iris.select_dtypes(include='number').corr()
correlation
```

```python
# `heatmap` permet la présentation en matrice
sns.heatmap(correlation)
```

```python
# Activer les annotatoins
sns.heatmap(correlation, annot = True)
```

_La couleur est claire si la corrélation est forte._

```python
# Mettre des couleurs chaudes si la corrélation est forte
sns.heatmap(correlation, annot = True, cmap = 'coolwarm')
```

```python
# Autres nuances
sns.heatmap(correlation, annot = True, cmap = 'viridis')
```

```python
sns.heatmap(correlation, annot = True, cmap = 'seismic')
```

```python
sns.heatmap(correlation, annot = True, cmap = 'flag')
```

```python
sns.heatmap(correlation, annot = True, cmap = 'coolwarm', linewidths = 3, linecolor = 'black')
```
