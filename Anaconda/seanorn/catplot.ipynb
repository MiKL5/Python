```python
import numpy   as np
import pandas  as pd
import seaborn as sns
```

```python
student = pd.read_csv('student.csv')
student.head()
```

# **La fonction `catplot()`**

Efficace pour travailler les données catégorielles, trace huit types de graphiques en modifiant les valeurs d'un paramètre dans `kind`.

```python
# La boxplot
sns.catplot(y = 'Note1', data = student, kind='box')
```

```python
sns.catplot(x = 'Note1', data = student, kind='box')
```

```python
sns.catplot(y = 'Note1', data = student, kind='boxen')
```

```python
sns.catplot(x = 'Note1', data = student, kind='boxen')
```

```python
sns.catplot(y = 'Note1', data = student, kind='violin')
```

```python
sns.catplot(x = 'Note1', data = student, kind='violin')
```

```python
sns.catplot(y = 'Note1', data = student, kind='strip')
```

```python
sns.catplot(x = 'Note1', data = student, kind='strip')
```

```python
sns.catplot(y = 'Note1', data = student, kind='point')
```

```python
sns.catplot(x = 'Note1', data = student, kind='point')
```

```python
sns.catplot(y = 'Note1', data = student, kind='count')
```

```python
sns.catplot(x = 'Note1', data = student, kind='count')
```
