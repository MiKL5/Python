# **Croisement de deux variables quantitatives**

`jointplot` ajoute des histogrammes pour chaque variable.

```python
sns.jointplot(y = 'Note3', x= 'absences', data = student)
```

```python
# faire une rÃ©grÃ©ssion linÃ©raire `reg` pour meiux comprendre le nuage de points
sns.jointplot(y = 'Note3', x= 'absences', data = student, kind = 'reg')
```

Plus il y a d'absence, plus la note est basse.

```python
sns.jointplot(y = 'Note3', x= 'Medu', data = student, kind = 'reg')
```

Si la mÃ¨re Ã  un meilleur niveau d'Ã©tude, les Ã©lÃ¨ves ont une meilleure note.

```python
sns.jointplot(y = 'Note3', x= 'Fedu', data = student, kind = 'reg')
```

Idem avec les pÃ¨res.

```python
# Avec des hexagones
sns.jointplot(y = 'Note3', x= 'Fedu', data = student, kind = 'hex')
```

```python
# Estimation de la densitÃ© en 2D
sns.jointplot(y = 'Note3', x= 'Fedu', data = student, kind = 'kde')
```
