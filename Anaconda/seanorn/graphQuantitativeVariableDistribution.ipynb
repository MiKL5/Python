# **Graphique d'une distribution de variable quantitative**

Les fonctions `displot` et `histplot` tracent les histogrammes avec une estimation  de la densité.
`distplot` sera supprimée dans une future version.

La fonction displot est la méthode recommandée pour visualiser des distributions _univariées_ ou _bivariées_ de données. Elle a été introduite dans la version 0.11.0 de Seaborn et remplace progressivement la fonction `distplot`, désormais dépréciée et prévue pour suppression dans la version 0.14.0). `displot` est plus flexible et puissante, permettant de créer des histogrammes, des courbes de densité (KDE), ainsi que des visualisations multi-facettes.

Pour résumer :
* displot est la fonction moderne et recommandée pour les plots de distribution.
* distplot est encore supportée, bien que dépréciée, et son usage est déconseillé dans les nouveaux projets.
* histplot est une autre fonction utilisée pour des histogrammes simples.

Il est conseillé d'utiliser `displot` pour les visualisations de distributions.

```python
# Créer l'histogramme avec displot
sns.displot(student.Note1)
```

```python
# Idem avec `histplot`
sns.histplot(student.Note1) # Illustration identique
```

```python
# Avec `displot` en 5 tranches
sns.displot(student.Note1, bins=5)
```

```python
sns.displot(student.Note1, bins=20, kde=True)# 20inttervales, plus la courbe
```

```python
# En boxplot horizontale
sns.boxplot(y='Note1', data=student)
```

```python
# En boxplot verticale
sns.boxplot(x='Note1', data=student)
```

```python
# Représenter la densité d'une variable quantitative
sns.violinplot(y='Note1', data=student)
```

```python
# Représenter la densité d'une variable quantitative à l'horizontale
sns.violinplot(x='Note1', data=student)
```

```python
# Représenter en nuage de points
sns.stripplot(y='Note1', data=student)
```

```python
sns.stripplot(x='Note1', data=student)
```

```python
# Représenter la moyenne de l'écart type
sns.pointplot(y='Note1', data=student)
```

Le point est la note moyenne. L'intervalle de confiance est entre 11,2 et 11,6.

```python
sns.pointplot(x='Note1', data=student)
```
