```python
import numpy      as np
import pandas     as pd
import seaborn    as sns
import matplotlib as mlt
```

# **Tableaux croisés dynamiques**

```python
passengerFlight = pd.read_csv('vols.csv')
passengerFlight.head()
```

```python
# Quelles sont les années de vols ?
passengerFlight['year'].unique()
```

```python
pf = passengerFlight.pivot_table(index = 'month', columns = 'year', values = 'passengers')
pf
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
sns.heatmap(pf, annot = True, cmap = 'coolwarm', linewidths = 3, linecolor = 'black')
plt.show()
```

```python
sns.heatmap(pf, cmap = 'coolwarm', linewidths = 3, linecolor = 'lightgray')
```
