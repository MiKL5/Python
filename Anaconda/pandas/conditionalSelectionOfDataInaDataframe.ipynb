# **Séléction conditionnelle des données d'un dataframe**

```python
notes = pd.DataFrame({
    'Français': {"Christophe": 0, "Carla": 5, "Caroline": 20, "Jean-Baptiste": 12, "Florence": 19, "Emma": 19},
    'Histoire': {"Christophe": 0, "Carla": 9, "Caroline": 20, "Jean-Baptiste": 18, "Florence": 20, "Emma": 19},
    'Maths': {"Christophe": 17, "Carla": 6, "Caroline": 20, "Jean-Baptiste": 2, "Florence": 16, "Emma": 19},
    'Anglais': {"Christophe": 1, "Carla": 14, "Caroline": 18, "Jean-Baptiste": 17, "Florence": 17, "Emma": 20},
    'Informatique': {"Christophe": 0, "Carla": 15, "Caroline": 19, "Jean-Baptiste": 17, "Florence": 18, "Emma": 11}
})

# Calcule la moyenne et l'ajoute en dernière colonne
notes['Moyenne'] = notes.mean(axis=1)
notes
```

```python
# En boulléen, vérifier les notes > 10
notes > 10
```

```python
notes_bool = notes > 10
notes_bool
```

```python
# Récupérer que les notes supérieures à 10
notes[notes_bool]
```

`NaN` pour _`Not a Number`_ s'affiche car le tableau ne prend pas ce qui est inférieur `<` à 10

```python
# Faire plus pratique
notes[notes > 10]
```

```python
notes['Informatique'] > 10
```

```python
# Qui n'a pas la moyenne
notes['Moyenne'] <= 10
```

Carla et Christophe ne l'ont pas

```python
# Tous les étudiants qui ont la moyenne
notes[notes['Moyenne'] >= 10]
```

> #### Les notes d'informatique des élève ayant au moins la moyenne en anglais

```python
notes_anglais = notes[notes['Anglais'] >= 10]
notes_anglais
```

```python
notes_anglais['Informatique']
```

> Deuxième méthode

```python
notes[notes['Anglais'] > 10]['Informatique']
```
